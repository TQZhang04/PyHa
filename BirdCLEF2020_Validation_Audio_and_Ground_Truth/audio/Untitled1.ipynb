```python


for file in os.listdir(audio_dir):
    print(file)
    try:
        filename,extension = file.split(".")
    except:
        continue
    for i in range(10):
        print(i)
        try:
            t1 = i * 60
            t2  = (i+1) * 60 
            t1 = t1 * 1000 #Works in milliseconds
            t2 = t2 * 1000
           
            rate, signal = wavfile.read(audio_dir + file)
            channel1 = signal

            newAudio = AudioSegment(
                channel1.tobytes(), 
                frame_rate=rate,
                sample_width=channel1.dtype.itemsize, 
                channels=1
            )
            newAudio = newAudio[t1:t2]
            newAudio.export(new_audio_dir+filename+"_"+str(i)+"."+extension, format="wav")
            print(new_audio_dir+filename+"_"+str(i)+"."+extension)
        except Exception as e:
            print(e)
            continue
```
